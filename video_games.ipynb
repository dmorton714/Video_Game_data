```python
# makes a data frame with critic score over 90 
critic_90 = df.query('Critic_Score >= 90')
critic_90 = critic_90.sort_values(by = 'Critic_Score', ascending=False)
critic_90
```

```python
# shows the platfor with critic scores above 90 
critic_90['Platform'].value_counts()

```

```python
critic_90['Platform'].value_counts().plot.bar()
```

```python
# user scores above 90
user_90 = df.query('User_Score >= 9')
user_90 = user_90.sort_values(by = 'User_Score', ascending=False)
user_90
```

```python
# shows the platfor with user scores above 90 
user_90['Platform'].value_counts()

```

```python
user_90['Platform'].value_counts().plot.bar()
```

```python
x1 = critic_90['Platform'].value_counts()
x2 = user_90['Platform'].value_counts()

comp = {
  "Critic": x1,
  "User": x2,
}
comp = pd.DataFrame(data=comp)

comp.plot.bar()
```
